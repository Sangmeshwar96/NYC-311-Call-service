```python
#complaint type and count table
majorcomplints=service311.dropna(subset=["Complaint Type"])
majorcomplints=service311.groupby("Complaint Type")
sortedComplaintType = majorcomplints.size().sort_values(ascending =False)
sortedComplaintType = sortedComplaintType.to_frame('count').reset_index()
sortedComplaintType
```

```python
sortedComplaintType.head(10)
```

```python
#pie graph of complaint type
sortedComplaintType = sortedComplaintType.head()
plt.figure(figsize=(5,5))
plt.pie(sortedComplaintType['count'],labels=sortedComplaintType["Complaint Type"], autopct="%1.2f%%")
plt.show()
```

```python
#group dataset of complaint type
groupedby_complainttype = service311.groupby('Complaint Type')
#10 major complaints in city 
groupedby_complainttype['Complaint Type'].value_counts().nlargest(10)
```
